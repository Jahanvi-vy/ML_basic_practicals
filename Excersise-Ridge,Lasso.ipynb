```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd

```

```python
dataset=pd.read_csv('Melbourne_housing_FULL.csv')
dataset.head()
```

```python
dataset.nunique()
```

```python
dataset.shape
```

```python
cols_to_use=['Suburb','Address','Type','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
```

```python
dataset=dataset[cols_to_use]
```

```python
dataset.isna().sum()
```

```python
cols_to_fill_zero=['Propertycount','Distance','Bedroom2','Bathroom','Car']
```

```python
dataset[cols_to_fill_zero]=dataset[cols_to_fill_zero].fillna(0)
dataset.isna().sum()
```

```python
dataset['Landsize']=dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea']=dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset
```

```python
#convert categorical feature into -encoding

dataset=pd.get_dummies(dataset,drop_first=True)
dataset.head()
```

```python
x=dataset.drop('Price',axis=1)
```

```python
y=dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression().fit(train_x,train_y)
```

```python
reg.score(test_x,test_y)#->ans 0.13
```

```python
reg.score(train_x,train_y)#->ans 0.68
```

```python
#this is condition of overfit 
# class sklearn.linear_model.Ridge(alpha=1.0, *, fit_intercept=True, copy_X=True, max_iter=None, tol=0.0001, solver='auto',)
```

Constant that multiplies the L2 term, controlling regularization strength.
fit_interceptbool, default=True
Whether to fit the intercept for this model. If set to false, no intercept will be used in calculations (i.e. X and y are expected to be centered).
Maximum number of iterations for conjugate gradient solver.
copy_Xbool, default=True
If True, X will be copied; else, it may be overwritten.

Regularization improves the conditioning of the problem and reduces the variance of the estimates. Larger values specify stronger regularization


```python
from sklearn import linear_model
lasso_reg=linear_model.Lasso(alpha=50,max_iter=100,tol=0.1)
lasso_reg.fit(train_x,train_y)
```

```python
lasso_reg.score(test_x,test_y)
```

```python
lasso_reg.score(train_x,train_y)
```

```python
#Ridge regression 
from sklearn.linear_model import Ridge 
ridge_reg=Ridge(alpha=50,max_iter=100,tol=0.1)
ridge_reg.fit(train_x,train_y)
```

```python
ridge_reg.score(test_x,test_y)
```

```python
ridge_reg.score(train_x,train_y)
```
